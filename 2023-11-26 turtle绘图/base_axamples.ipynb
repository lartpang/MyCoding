```python
import turtle as tt
import numpy as np
```

```python
tt.screensize(600, 600, "white")
tt.pensize(5)
tt.color('black')
radius = 20
tt.begin_fill()
tt.fillcolor('red')
tt.penup()
tt.goto(0, -6*radius)
tt.pendown()
tt.circle(6*radius)
tt.end_fill()
tt.penup()
tt.begin_fill()
tt.fillcolor('black')
tt.goto(3*radius, 2*radius)
tt.pendown()
tt.right(90)
for i in range(4):
    tt.forward((5+3**0.5)*radius)
    tt.right(150)
    tt.forward(2*radius)
    tt.left(60)
tt.end_fill()
tt.penup()
tt.goto(radius, 0)
tt.right(180)
tt.pencolor('red')
tt.pendown()
tt.circle(radius)
tt.done()
```

```python
tt.screensize(600, 600, "white")
tt.pensize(20)
tt.color('black')
r = 30

tt.penup()
tt.goto(0, 3*r)
tt.pendown()
tt.circle(-3*r, 45)
tt.left(90)
tt.fd(2*r)
tt.left(135)
tt.penup()
tt.goto(0, 3*r)
tt.pendown()
tt.circle(3*r, 180)
tt.circle(2*r, 180)
tt.circle(r, 180)
tt.left(180)
tt.penup()

tt.goto(0, -3*r)
tt.pendown()
tt.fd(3 * np.sqrt(3) * r)
tt.right(120)
tt.fd(3 * np.sqrt(3) * r)
tt.hideturtle()  # 隐藏箭头
tt.done()
```

```python
import math

tt.screensize(600, 600, "white")
len_line = 60
tt.pensize(10)
tt.color('blue')
tt.speed('fast')

tt.penup()
tt.goto(-30*math.sqrt(3), -30)
tt.pendown()
tt.left(90)

for i in range(6):
    tt.fd(len_line)
    tt.left(60)
    tt.fd(len_line)
    tt.left(60)
    tt.fd(len_line)
    tt.right(180)
    tt.fd(len_line)
    tt.left(60)
    tt.fd(len_line)
    tt.right(180)
    tt.fd(len_line)
    tt.right(120)
    tt.fd(len_line)
    tt.left(60)
    tt.fd(len_line)
    tt.right(180)
    tt.fd(len_line)
    tt.left(60)
    tt.fd(len_line)
    tt.right(180)
    tt.fd(len_line)
    tt.left(60)
    tt.fd(2 * len_line)
    tt.left(60)
tt.penup()
tt.done()
```

```python
tt.screensize(600, 600, "white")
len_line = 60
tt.pensize(10)
tt.color('blue')
tt.speed('fast')

tt.penup()
tt.goto(-30*np.sqrt(3), -30)
tt.pendown()
tt.left(90)

position = []
for i in range(6):
    position.append(tt.position())
    tt.fd(len_line)
    tt.right(60)

tt.left(120)

for i in range(6):
    tt.penup()
    tt.goto(position[i])
    tt.pendown()
    tt.fd(2*len_line)
    tt.right(60)
    tt.fd(len_line)
    tt.right(180)
    tt.fd(len_line)
    tt.right(60)
    tt.fd(len_line)
    tt.right(60)
    tt.penup()
    tt.goto(position[i])
    tt.pendown()
    tt.fd(len_line)
    tt.right(60)
    tt.fd(len_line)
    tt.right(180)
    tt.fd(len_line)
    tt.right(60)
    tt.fd(len_line)
    tt.right(60)
    tt.penup()
    tt.goto(position[i])
    tt.right(60)

tt.done()
```
